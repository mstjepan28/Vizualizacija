```python
import pandas as pd
import json

covid_data = pd.read_csv('download.csv')
```

```python
covid_data
```

```python
countries = pd.read_csv('countries.csv')

country_cords = {}
for i, row in countries.iterrows():
    key = row['country']
    
    lat = row['latitude']
    lon = row['longitude']
    value = [lat, lon]
    
    country_cords[key] = value
    
with open('country_dict.json', 'w') as fp:
    json.dump(country_cords, fp)
```

```python
with open('country_dict.json', 'r') as fp:
    country_cords = json.load(fp)    
```

```python
def get_cords(country_code, country_code_2):
    coordinates = None
    
    try:
        coordinates = country_cords[country_code]
    except KeyError:
        try:
            coordinates = country_cords[country_code_2]
        except KeyError:
            print(country_code, country_code_2)
        
    return coordinates

cur_country = None
cords = None
cur_code = None
cur_cases = 0
cur_deaths = 0

countries = []
code = []
cases = []
deaths = []
lat =  []
long = []

for index, row in covid_data.iterrows():
    if cur_country != row['countriesAndTerritories']:       
        if cords:
            countries.append(cur_country)
            code.append(cur_code)
            cases.append(cur_cases)
            deaths.append(cur_deaths)
            lat.append(cords[0])
            long.append(cords[1])
         
        cur_country = row['countriesAndTerritories']
        cur_code = row['countryterritoryCode']
        if not cur_code:
            cur_code = row['geoId']
        cur_cases = 0
        cur_deaths = 0        
        cords = get_cords(row['geoId'], row['countryterritoryCode'])     
        if not cords:
            #print(cur_country)
            continue
    
    cur_cases += row['cases']
    cur_deaths += row['deaths']

if cords:
    countries.append(cur_country)
    code.append(cur_code)
    cases.append(cur_cases)
    deaths.append(cur_deaths)
    lat.append(cords[0])
    long.append(cords[1])
    
print("Done")
```

```python
filtered_data = pd.DataFrame({
    'country': countries,
    'code': code,
    'n': cases,
    'n1': deaths,
    'Latitude': lat,
    'Longitude': long
})
filtered_data
```

```python
filtered_data.to_csv('COVID-19_Data.csv', index=False)
```
