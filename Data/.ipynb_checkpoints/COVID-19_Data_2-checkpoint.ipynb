```python
import pandas as pd
import datetime

covid_podaci = pd.read_csv('download.csv')
```

```python
#Dict koristimo za zbrajanje slucajeva 
rijecnik_podataka = {}
#rijecnik = {
#     datum: lista[n_slucajeva, n_smrti]
#    '01/01/2020': [10, 1]
#}

for index, red in covid_podaci.iterrows():
    datum = red['dateRep']
    
    trenutni_slucajevi = red['cases']
    trenutne_smrti = red['deaths']
    
    #Ako ne postoji taj datum u rijecniku, stvori da s danom vrijednoscu
    try:
        rijecnik_podataka[datum][0] += trenutni_slucajevi
    except:
        rijecnik_podataka[datum] = [trenutni_slucajevi, 0]
        
    try:
        rijecnik_podataka[datum][1] += trenutne_smrti
    except:
        rijecnik_podataka[datum] = [0,trenutne_smrti]

print("Done")
```

```python
#Prima parametar formata [[n_slucajeva, n_smrti], [10, 2],...]
#Vraca tuple ([n_slucajeva],[n_smrti])
def razdvoji(lista):
    n_slucajeva = []
    n_smrti = []
    
    for element in lista:
        n_slucajeva.append(element[0])
        n_smrti.append(element[1])
    
    return (n_slucajeva, n_smrti)

#Iz rijecnika se izvlace podaci, datumi, broj slucaja, broj umrlih
datumi = list(rijecnik_podataka.keys())
podaci = list(rijecnik_podataka.values())

#Oblikovanje podataka
datumi = [datetime.datetime.strptime(datum, '%d/%m/%Y').date() for datum in datumi]
n_slucajevi, n_smrti = razdvoji(podaci)

#Stvaranje novog dataframea sa obradenim podacima
novi_df = pd.DataFrame({
    'datum': datumi,
    'slucajevi': n_slucajevi,
    'smrti': n_smrti
})
novi_df.sort_values('datum', ascending=True, inplace=True)
novi_df.reset_index(drop = True, inplace = True)

novi_df
```

```python
novi_df.to_csv('COVID-19_Data_2.csv', index=False)
```
