```python
import pandas as pd
import json

covid_data = pd.read_csv('download.csv')
```

```python
covid_data
```

```python
countries = pd.read_csv('countries.csv')

country_cords = {}
for i, row in countries.iterrows():
    key = row['country']
    
    lon = row['latitude']
    lat = row['longitude']
    value = [lat, lon]
    
    country_cords[key] = value
    
with open('country_dict.json', 'w') as fp:
    json.dump(country_cords, fp)
```

```python
with open('country_dict.json', 'r') as fp:
    country_cords = json.load(fp)    
```

```python
def find_max(list):
    max = 0
    for num in list:
        if max < num:
            max = num
    return max

def get_cords(country_code, country_code_2):
    coordinates = None
    
    try:
        coordinates = country_cords[country_code]
    except KeyError:
        try:
            coordinates = country_cords[country_code_2]
        except KeyError:
            print(country_code, country_code_2)
        
    return coordinates

cur_country = None
cords = None
cur_cases = []
cur_deaths = []

countries = []
cases = []
deaths = []
lat =  []
long = []

for index, row in covid_data.iterrows():
    if cur_country != row['countriesAndTerritories']:       
        if cords:
            countries.append(cur_country)
            cases.append(find_max(cur_cases))
            deaths.append(find_max(cur_deaths))
            lat.append(cords[0])
            long.append(cords[1])
         
        cur_country = row['countriesAndTerritories']
        cur_cases = []
        cur_deaths = []        
        cords = get_cords(row['geoId'], row['countryterritoryCode'])     
        if not cords:
            print(cur_country)
            continue
        
    cur_cases.append(row['cases'])
    cur_deaths.append(row['deaths'])

if cords:
    countries.append(cur_country)
    cases.append(find_max(cur_cases))
    deaths.append(find_max(cur_deaths))
    lat.append(cords[0])
    long.append(cords[1])
    
print("Done")
```

```python
filtered_data = pd.DataFrame({
    'Country': countries,
    'Cases': cases,
    'Deaths': deaths,
    'Latitude': lat,
    'Longitude': long
})
filtered_data
```

```python
filtered_data.to_csv('COVID-19_Data.csv')
```
